```python
import requests
import json

# --- Configuration ---
# Replace with your actual Polymarket API Key
API_KEY = 'YOUR_POLYMARKET_API_KEY'
# Replace with the actual Polymarket API base URL for data
BASE_URL = 'https://gamma-api.polymarket.com/' # This is a hypothetical URL

# --- Function to fetch popular contract IDs ---
def get_popular_contract_ids(limit=5):
    headers = {
        # 'Authorization': f'Bearer {API_KEY}',
        # 'Content-Type': 'application/json'
    }
    # Hypothetical endpoint for popular markets. Adjust according to Polymarket API docs.
    # It's assumed this endpoint returns a list of market objects, each with an 'id'.
    endpoint = f'{BASE_URL}events?active=true&closed=false&limit={limit}&order=volume'

    try:
        print(f"Attempting to fetch popular contract IDs from: {endpoint}")
        response = requests.get(endpoint, headers=headers)
        response.raise_for_status()  # Raise an HTTPError for bad responses (4xx or 5xx)
        data = response.json()

        # Assuming the response is a list of market objects and each has an 'id' field
        contract_ids = [market['id'] for market in data if 'id' in market]
        print(f"Fetched {len(contract_ids)} popular contract IDs.")
        return contract_ids
    except requests.exceptions.HTTPError as http_err:
        print(f"HTTP error occurred fetching popular contracts: {http_err}")
        print(f"Response content: {response.text}")
    except requests.exceptions.ConnectionError as conn_err:
        print(f"Connection error occurred: {conn_err}")
    except requests.exceptions.Timeout as timeout_err:
        print(f"Timeout error occurred: {timeout_err}")
    except requests.exceptions.RequestException as req_err:
        print(f"An error occurred: {req_err}")
    return []

# --- Function to fetch data for a single contract ---
def get_contract_data(contract_id):
    headers = {
        # 'Authorization': f'Bearer {API_KEY}',
        # 'Content-Type': 'application/json'
    }
    # Adjust the endpoint path based on actual Polymarket API documentation
    endpoint = f'{BASE_URL}events/{contract_id}' # Hypothetical endpoint

    try:
        response = requests.get(endpoint, headers=headers)
        response.raise_for_status()  # Raise an HTTPError for bad responses (4xx or 5xx)
        return response.json()
    except requests.exceptions.HTTPError as http_err:
        print(f"HTTP error occurred: {http_err} for contract {contract_id}")
        print(f"Response content: {response.text}")
    except requests.exceptions.ConnectionError as conn_err:
        print(f"Connection error occurred: {conn_err}")
    except requests.exceptions.Timeout as timeout_err:
        print(f"Timeout error occurred: {timeout_err}")
    except requests.exceptions.RequestException as req_err:
        print(f"An error occurred: {req_err}")
    return None

# --- Main execution ---
all_contracts_data = []
print("Fetching data for Polymarket contracts...")

# Get popular contract IDs dynamically
CONTRACT_IDS = get_popular_contract_ids(limit=5)

if not CONTRACT_IDS:
    print("Could not fetch popular contract IDs. Please check API key, base URL, and endpoint for popular markets.")
else:
    for i, contract_id in enumerate(CONTRACT_IDS):
        print(f"\nFetching data for contract {i+1}/{len(CONTRACT_IDS)}: {contract_id}")
        data = get_contract_data(contract_id)
        if data:
            all_contracts_data.append(data)
            print(f"Successfully fetched data for {contract_id}")
            # Optional: print a snippet of the data
            # print(json.dumps(data, indent=2))
        else:
            print(f"Failed to fetch data for {contract_id}")

print("\n--- Download Complete ---")
print(f"Total contracts data fetched: {len(all_contracts_data)}")

# You can now process or save 'all_contracts_data'
# Example: print the first fetched contract's data
if all_contracts_data:
    print("\nFirst contract data example:")
    print(json.dumps(all_contracts_data[0], indent=2))
```
